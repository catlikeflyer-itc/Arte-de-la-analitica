```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("datos/nutrients_csvfile.csv")
df.head()
```

```python
df.Category.unique()
```

```python
df = df[["Food", "Grams", "Calories", "Category"]]
df = df.loc[df["Category"].isin(["Vegetables", "Seeds and Nuts", "Breads. cereals. fastfood.grains"])]

sns.scatterplot(data=df, x="Grams", y="Calories", hue=df.Category, palette="coolwarm_r")
```

```python
X = df[["Grams", "Calories"]]
km = KMeans(n_clusters=12, n_init=12, init="random", random_state=42)
km.fit(X)
y_kmeans = km.predict(X)
y_kmeans
```

```python
sns.scatterplot(data=df, x="Grams", y="Calories", hue=df.Category, palette="coolwarm_r")
centers = km.cluster_centers_

plt.scatter(centers[:,0], centers[:, 1], c="black", s=100, alpha=0.6)
```
